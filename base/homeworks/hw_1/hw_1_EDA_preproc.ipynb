## Найдите и исправьте ошибки в данных

```python
nan_count = dict()
for col in df:
    nan_count[col] = (1 - len(df[df[col].notnull()][col]) / df.shape[0]) * 100
```

```python
nan_count = {k: v for k, v in sorted(nan_count.items(), key=lambda item: item[1], reverse=True)}
```

```python
plt.figure(figsize=(15,7))
plt.title("Процентное содержание Nan")
plt.bar(nan_count.keys(), nan_count.values())
plt.tick_params(axis='x', rotation=35)
pass
```

```python
df.loc[df['age'].isna(), 'age'] = 15
df.loc[df['address'].isna(), 'address'] = 'U'

# Мне кажется тут опечатки, поэтому так и правлю
df.loc[df['age'] == 116, 'age'] = 16
df.loc[df['age'] == 151, 'age'] = 15
df.loc[df['age'] == 161, 'age'] = 16
df.loc[df['age'] == 181, 'age'] = 18
df.loc[df['age'] == -1, 'age'] = 15 # этого просто до среднего

df.loc[(df['sex']=='D') | (df['sex']=='C') | (df['sex']=='A'), 'sex'] = "F"
df.loc[df['sex']=='B','sex'] = 'M'
```

Сложные признаки - это комбинации существующих (линейных или нелинейных).
Это нужно для того, чтобы самостоятельно добавить нелинейность в алгоритм.

Приведите не менее двух признаков.

```python
health_dacl = df['health'] * df['Dalc']
age_studytm = (df['age'] - min(df['age']))/(max(df['age']) - min(df['age'])) * df['studytime'] 
fail_dalc   = df['failures'] * df['Dalc']

data = df.iloc[:,:-1]
data['health_dacl'] = health_dacl
data['age_studytm'] = age_studytm
data['fail_dalc'] = fail_dalc 
data['Target'] = df['G3']
```
